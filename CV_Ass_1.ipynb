```python
import cv2 as cv
from matplotlib import pyplot as plt
import numpy as np

%matplotlib inline
```

```python
img = cv.imread('/content/low contrast.png')
```

setelah di import. 

kita konversi gambar kita menjadi grayscale terlebih dahulu

```python
grayimg = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
```

Lalu saya menggunakan fitur equalize histogram yang ada di opencv

```python
eqhist = cv.equalizeHist(grayimg)
```

```python
output = [grayimg,eqhist]
```

```python
titles = ['original image','equlize']
```

```python
for i in range(2):
    plt.subplot(1,2,i+1)
    plt.imshow(output[i],cmap='gray')
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
```

```python
for i in range(2):
    plt.figure(figsize=(20,5))
    plt.subplot(1,2,i+1)
    plt.hist(output[i].ravel(), 256 ,[0,256])
    # plt.legend(loc = 'upper right')
    plt.xlabel('Intensity')
    plt.ylabel('Quantity')
plt.show()
```

Untuk hasil equalizehist dapat kita lihat bahwa contrast rangenya menjadi lebih rata yaitu dari 0 hingga 250 walaupun lebih renggang karena keterbatasan data yang dimiliki dari input
