```python
import cv2 as cv
from matplotlib import pyplot as plt
import numpy as np

%matplotlib inline
```

```python
img = cv.imread('01-01.PNG',0)
```

setelah di import. 

saya menggunakan gaussian blur sesuai dengan yang di contohkan pada ppt tetapi bedanya dengan low pass, image kita kurangin dengan low pass yang telah ditambahkan 127( setengah dari 255)

```python
# gaussian = cv.GaussianBlur(img, (31,31), 0)
hpf = img - cv.GaussianBlur(img, (51, 51), 0)+127
```

Lalu saya hanya membandingkan asli dengan setelah di filter

```python
output = [img,hpf]
```

```python
titles = ['original image','high pass gaussian']
```

```python
for i in range(2):
    plt.subplot(1,2,i+1)
    plt.imshow(output[i],cmap='gray')
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
```

```python
f = np.fft.fft2(img)
fshift = np.fft.fftshift(f)
magnitude_spectrum = 20*np.log(np.abs(fshift))
plt.subplot(121),plt.imshow(img, cmap = 'gray')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(magnitude_spectrum, cmap = 'gray')
plt.title('Magnitude Spectrum'), plt.xticks([]), plt.yticks([])
plt.show()
```

Gambarnya memiliki low spot lagi di pinggir, tetapi karena mau mengambil high pass maka setelah di ambil low spotnya, kita tambahkan 127 agar menjadi high pass

```python
f = np.fft.fft2(hpf)
fshift = np.fft.fftshift(f)
magnitude_spectrum = 20*np.log(np.abs(fshift))
plt.subplot(121),plt.imshow(hpf, cmap = 'gray')
plt.title('Gaussian Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(magnitude_spectrum, cmap = 'gray')
plt.title('Magnitude Spectrum After Gaussian'), plt.xticks([]), plt.yticks([])
plt.show()
```

High Spot yang ada di tengah yang awalnya terang sekarang menjadi hilang
