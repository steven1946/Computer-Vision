```python
import cv2 as cv
from matplotlib import pyplot as plt
import numpy as np

%matplotlib inline
```

```python
img = cv.imread('01-01.PNG')
```

setelah di import. 

saya menggunakan gaussian blur sesuai dengan yang di contohkan pada ppt

```python
gaussian = cv.GaussianBlur(img, (31,31), 0)
```

Lalu saya hanya membandingkan asli dengan setelah di filter

```python
output = [img,gaussian]
```

```python
titles = ['original image','low pass gaussian']
```

```python
for i in range(2):
    plt.subplot(1,2,i+1)
    plt.imshow(output[i],cmap='gray')
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
```
