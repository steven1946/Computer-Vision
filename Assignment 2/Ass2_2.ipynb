```python
import cv2 as cv
from matplotlib import pyplot as plt
import numpy as np

%matplotlib inline
```

```python
img = cv.imread('01-01.PNG')
```

setelah di import. 

saya menggunakan gaussian blur sesuai dengan yang di contohkan pada ppt tetapi bedanya dengan low pass, image kita kurangin dengan low pass yang telah ditambahkan 127( setengah dari 255)

```python
# gaussian = cv.GaussianBlur(img, (31,31), 0)
hpf = img - cv.GaussianBlur(img, (51, 51), 0)+127
```

Lalu saya hanya membandingkan asli dengan setelah di filter

```python
output = [img,hpf]
```

```python
titles = ['original image','high pass gaussian']
```

```python
for i in range(2):
    plt.subplot(1,2,i+1)
    plt.imshow(output[i],cmap='gray')
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
```
