```python
import cv2 as cv
from matplotlib import pyplot as plt
import numpy as np

%matplotlib inline
```

```python
img = cv.imread('01-01.PNG', 0)
```

setelah di import. 

saya menggunakan gaussian blur sesuai dengan yang di contohkan pada ppt

```python
gaussian = cv.GaussianBlur(img, (31,31), 0)
```

Lalu saya hanya membandingkan asli dengan setelah di filter

```python
output = [img,gaussian]
```

```python
titles = ['original image','low pass gaussian']
```

```python
for i in range(2):
    plt.subplot(1,2,i+1)
    plt.imshow(output[i],cmap='gray')
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
```

```python
f = np.fft.fft2(img)
fshift = np.fft.fftshift(f)
magnitude_spectrum = 20*np.log(np.abs(fshift))
plt.subplot(121),plt.imshow(img, cmap = 'gray')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(magnitude_spectrum, cmap = 'gray')
plt.title('Magnitude Spectrum'), plt.xticks([]), plt.yticks([])
plt.show()
```

Low spotnya ada di pinggir gambar

```python
f = np.fft.fft2(gaussian)
fshift = np.fft.fftshift(f)
magnitude_spectrum = 20*np.log(np.abs(fshift))
plt.subplot(121),plt.imshow(gaussian, cmap = 'gray')
plt.title('Gaussian Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(magnitude_spectrum, cmap = 'gray')
plt.title('Magnitude Spectrum After Gaussian'), plt.xticks([]), plt.yticks([])
plt.show()
```

Setelah di apply gaussian, bagian pinggir jadi dibuang karena hal tersebut adalah tempat low spotnya
